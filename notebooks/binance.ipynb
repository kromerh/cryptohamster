```python
import json
import requests

url = 'https://api.binance.com/api/v3/ticker/price'
url = 'https://www.binance.com/api/v3/ticker/price?symbol=ANKRTUSD'
# requesting data from url
data = requests.get(url)  
data = data.json()
d = {}
for item in data:
    symbol = item['symbol']
    if symbol.endswith('TUSD'):
        d[item['symbol']] = float(item['price'])
```

```python
marklist = sorted(d.items(), key=lambda x:x[1], reverse=True)
sortdict = dict(marklist)
print(sortdict)
```

```python
import sys
from datetime import datetime
import os
import pandas as pd

# Add path to source files
sys.path.insert(1, os.path.join(sys.path[0], '..'))

from src.python.utils import load_credentials
```

```python
FULL_PATH_TO_CREDENTIALS = './../credentials_API.cred'
api_key, api_secret = load_credentials(filepath=FULL_PATH_TO_CREDENTIALS)
client = Client(api_key, api_secret)
```

```python
timestamp = client._get_earliest_valid_timestamp('BTCUSDT', '1h')
# request historical candle (or klines) data
bars = client.get_historical_klines('BTCUSDT', '1h', timestamp, limit=1000)
```

```python
# delete unwanted data - just keep date, open, high, low, close
for line in bars:
    line[0] = datetime.fromtimestamp(int(line[0])/1000)
    del line[5:]
```

```python
btc_df = pd.DataFrame(bars, columns=['date', 'open', 'high', 'low', 'close'])
btc_df.set_index('date', inplace=True)
```

```python
def myComp(list1, list2):
  section = list(set(list1).intersection(list2))
  return ((len(section)==0), section)

myComp(['a', 'b'], ['a', 'b'])
```

```python
def test(mysql_kwargs):
    expected_keys = ['host', 'user', 'password', 'db', 'port']
    expected_keys.sort()
    check_list = list(mysql_kwargs.keys())
    check_list.sort()

    try:
        assert expected_keys == check_list
    except AssertionError as e:
        raise ValueError(f"User input invalid. Not all mysql keywords in the keyword list.")
    
    return mysql_kwargs


input_d = {'host': '1', 'user': '2', 'password': '3', 'db': '4', 'port': '5'}
input_d2 = {'host': '1', 'user': '2', 'password': '3', 'db': '4', 'port': '5', 'another': '44'}
input_d1 = {'user': '2', 'password': '3', 'db': '4', 'port': '5'}
test(input_d)
```
